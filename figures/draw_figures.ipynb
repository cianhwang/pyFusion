```python
from matplotlib.ticker import MaxNLocator
T = [2, 3, 4]
random = [2.856, 2.642, 2.604]

greedy = [1.440, 1.414, 1.323]
ours = [1.427, 1.398, 1.293]

# plt.plot(T, random)
fig, ax = plt.subplots()
ax.plot(T, greedy, label = 'greedy')
ax.plot(T, ours, label = 'ours')
ax.set_xlabel('T')
ax.set_ylabel('MSE x1000')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.legend()
plt.savefig('greedy_vs_ours.png')
```
